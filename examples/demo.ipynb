# MD-States demo

```python
from IPython.display import Image

import mdstates
```

# Create an `mdstates.Network` object instance and add a replica to it.

```python
net = mdstates.Network()

# Requires a topology file even if the topology changes.
# The first frame of the trajectory as a .pdb file works well.
net.add_replica("example.xyz", "topology.pdb")
```

# Build the contact matrix of all frames in the trajectory.

```python
net.generate_contact_matrix()
```

# Use the Viterbi algorithm to smooth the trajectory contact matrices.
This step takes the longest time since it has to loop over every single frame.

```python
net.decode()
```

# Draw the network of all the replicas
By default, it saves the final image file as `overall.png`

The network is generated by `Graphviz` and the file `input.dot` is the input file that generates this specific network image

```python
net.draw_overall_network()
```
